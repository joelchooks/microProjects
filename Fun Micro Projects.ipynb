### Mini Bank Account Simulation

```python
class BankAccount():
    #This is an app to show bank balance, makes deposits and withdraws
    
    def __init__(self, balance=0.00):
        self.balance = balance
        
    def showBalance(self):
        print(f'Your balance is {self.balance}')
    
    def makeDeposit(self):
        amount = float(input("How much would you like to deposit?"))
        self.balance += amount
        print(f'DEPOSIT SUCCESSFUL! Your new balance is {self.balance}')
        
    def makeWithdrawal(self):
        amount = float(input("How much would you like to withdraw?"))
        if amount > self.balance:
            print(f'Sorry, you have insufficient funds to make this withdrawal. Your balance is {self.balance}')
        else:
            self.balance -= amount
            print(f'WITHDRAWAL SUCCESSFUL! Your new balance is {self.balance}')
        
```

```python
my_acct = BankAccount()
```

```python
my_acct.makeDeposit()
```

```python
my_acct.showBalance()
```

```python
my_acct.makeWithdrawal()
```

### Area of a Circle

```python
import math

class returnArea(object):
    #This app displays the area of a circle given its radius
    
    def __init__(self, value):
        self.value = value
    
    def area(self):
        ans = math.pi*pow(self.value, 2)
        return round(ans, 3)
```

```python
abc = returnArea(8)
```

```python
abc.area()
```

### Caesar Cipher

```python
class EncryptDecrypt(object):
    
    #This app encrypts and decrypts messages passed to it
    
    abc = "abcdefghijklmnopqrstuvwxyz"
    
    def shift_text(k):
        return k%26
        
    def encrypt(self):
        output = ''
        text = str(input("Please enter text to encrypt: "))
        num  = int(input("Please number you would encrypt by: "))
        text = text.lower()
        for u in text:
            if u not in abc:
                output = output + u
            else:
                alpha_index = abc.find(u)
                output = output + abc[shift_text(alpha_index + num)]
        return output
                
    def decrypt(self):
        output = ''
        text = str(input("Please enter text to decrypt: "))
        num  = int(input("Please number you would decrypt by: "))
        text = text.lower()
        for u in text:
            if u not in abc:
                output = output + u
            else:
                alpha_index = abc.find(u)
                output = output + abc[shift_text(alpha_index - num)]
        return output
```

```python
enc = EncryptDecrypt()
```

```python
enc.encrypt()
```

```python
enc.decrypt()
```

### Randam Walk Simulation

```python
import matplotlib.pyplot as plt
from random import choice
plt.rc('figure', figsize = (14,6))
```

```python
class NewWalk():
    #Create class for random walk
    
    def __init__(self, step_num):
        self.step_num = step_num
        
        
    def rand_walk(self):
    #Define function for random walk
    
        step_option = [1, -1]
        step_choice = choice(step_option)  
        walk = []
        walk.append(step_choice)
        for i in range(1,self.step_num):
            next_step = walk[-1] + choice(step_option)
            walk.append(next_step)   
        return walk

    def plot_walk(self):

        #Define a function to plot random walk
        plt.plot(rand_walk(self.step_num))
        plt.xlabel("Number of walks")
        plt.ylabel("Distance from Origin")
        plt.title("Our Random Walk")
        plt.show()
        
    def plot_many_walks(self, num):
        lab_list = list(range(1,num+1))

        for x in range(0,num):
            plt.plot(rand_walk(self.step_num), label = "Plot Number " + str(lab_list[x]))
            plt.xlabel("Number of walks")
            plt.ylabel("Distance from Origin")
            plt.title("Our Random Walk(s)")
            plt.legend(loc = "lower left")
        plt.show()
    
```

```python
ab = NewWalk(1000)
```

```python
ab.plot_many_walks(5)
```
